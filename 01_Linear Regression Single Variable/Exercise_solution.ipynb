Predict canada's per capita income in year 2020 using canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020 and 2022.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
plt.scatter(df.year,df.percapitaincome,color="r",marker=".")
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']].values,df.percapitaincome)
```

```python
reg.predict([[2022]])
```

```python
plt.scatter(df.year,df.percapitaincome,color="r",marker=".")
plt.plot(df.year, reg.predict(df[['year']].values))
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.show()
```
