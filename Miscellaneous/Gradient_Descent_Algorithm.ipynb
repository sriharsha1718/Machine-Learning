```python
#Importing necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import time
```

**Define the function**


*   **f(x) = 2x^2 + 4x + 5**



```python
X = np.arange(-100,100)
Y = (2*(X**2)) + (4*X) + 2
print(X)
```

```python
print(Y)
```

```python
#Plot the function
plt.title("2x^2 + 4x + 5")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.plot(X,Y)
plt.show()
```

**Gradient Descent Algorithm**

```python
plt.title("f(x) = 2x^2 + 4x + 5")
plt.xlabel("X")
plt.ylabel("Y")

#Intilialize a random value to x
x = -90

#Initialize the learning rate
lr = 0.1

plt.plot(X,Y)
initial = time.time()
for i in range(1000):
  gradient = (4*x) + 4
  x = x - (lr*gradient)
  y = (2*(x**2)) + (4*x) + 5
  plt.scatter(x,y)
final = time.time()
print("The Elapsed Time is:",(final-initial))
plt.grid()
plt.show()
```

```python
#Tweaking The learning rate
plt.title("f(x) = 2x^2 + 4x + 5")
plt.xlabel("X")
plt.ylabel("Y")

#Intilialize a random value to x
x = -90

#Initialize the learning rate
lr = 0.01

plt.plot(X,Y)
initial = time.time()
for i in range(1000):
  gradient = (4*x) + 4
  x = x - (lr*gradient)
  y = (2*(x**2)) + (4*x) + 5
  plt.scatter(x,y)
final = time.time()
print("The Elapsed time: ", (final-initial))
plt.grid()
plt.show()
```
