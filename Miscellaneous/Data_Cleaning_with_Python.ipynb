**Insightful Data Extraction**

```python
yearGroup = df.groupby('arrival_date_year')['arrival_date_year'].count()
print(yearGroup)
```

```python
mostMonth = df.groupby('arrival_date_month')['arrival_date_month'].count()
print(mostMonth)
```

```python
sum_columns = df[['adults', 'children', 'babies']].sum()
print(sum_columns)

print("The Adults Column Sum >>", sum_columns['adults'])
print("The Children Column Sum >>", sum_columns['children'])
print("The Babies Column Sum >>", sum_columns['babies'])
```

**Total Repeated Guest**

```python
reptGuestMonth = df.groupby('arrival_date_month')['is_repeated_guest'].sum().idxmax()
reptGuest = df.groupby('arrival_date_month')['is_repeated_guest'].sum().max()
print("Repeated Guests came in the month of",reptGuestMonth)
print("The No of Repeated Guests are",reptGuest)
```

```python
LreptGuestMonth = df.groupby('arrival_date_month')['is_repeated_guest'].sum().idxmin()
LreptGuest = df.groupby('arrival_date_month')['is_repeated_guest'].sum().min()
print("Least Repeated Guests came in the month of",LreptGuestMonth)
print("The No of Least Repeated Guests are",LreptGuest)
```
